```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder \
.master("local[4]") \
.appName("DataExplore") \
.config("spark.executor.memory","4g") \
.config("spark.driver.memory","4g") \
.getOrCreate()
```

```python
adult_train_df = spark.read \
.option("header","true") \
.option("inferSchema","True") \
.option("sep",",") \
.csv("/home/batuhan/Desktop/adult.data") 
```

```python
adult_test_df = spark.read \
.option("header","true") \
.option("inferSchema","True") \
.option("sep",",") \
.csv("/home/batuhan/Desktop/adult.test") 
```

```python
adult_test_df.show(100)
```

```python
adult_train_df.limit(10).toPandas().head()
```

```python
adult_test_df.limit(10).toPandas().head()
```

# 2 tane Pandas DateFrame birleitşrme 

```python
adult_whole_df = adult_train_df.union(adult_test_df)
adult_whole_df.limit(5).toPandas().head()
```

```python
print("train satır sayısı: ", adult_train_df.count())
print("test satır sayısı :", adult_test_df.count())
print("whole satır sayısı : ", adult_whole_df.count())
```

# KATEGORİK DEĞİŞKENLERİN İNCELENMESİ

```python
# Kategorik Değişkenlerin GroupBy() ile incelemek daha çok bilgi verir. Heef değişken dahil toplam tüm
# değişkenler var
```

```python
from pyspark.sql import functions as F
```

```python
adult_whole_df.groupBy(F.col("workclass")) \
.agg({"*":"count"}) \
.toPandas().head()
```

```python
# Biz Ne Yapmış Olduk ?
# Yukarıdaki groupby komutu ile workclass kategorik değişkenindeki tü string ifadeleri saydırdık eper mesela
# ? veya yanlış yazılmış veri olsa idi workclass sütununda onları da sayacaktı ve bize 1 tane State-goc gibi yanlış
# yazılmış dğeişkeni gösterecekti
# Bunu tüm kategorik değişkenlerde inceleyebilirsin
```
